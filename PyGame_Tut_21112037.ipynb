```python
import pygame
import sys
import random

# Initialize Pygame
pygame.init()
pygame.mixer.init()

# Get the screen dimensions
screen_info = pygame.display.Info()
screen_width = screen_info.current_w
screen_height = screen_info.current_h

# Calculate the window size (a quarter of the screen)
window_width = screen_width // 2
window_height = screen_height // 2

# Create the game window
game_display = pygame.display.set_mode((window_width, window_height))

# Set a title for your game window
pygame.display.set_caption("BlastBall Pro Max")

# Load the space background image
background_image = pygame.image.load("space_background.jpeg")

# Scale the background image to match the window size
background_image = pygame.transform.scale(background_image, (window_width, window_height))


# ---------------------------------------------------------------- Pad Initialization : ---------------------------------------------------
# Initialize pad movement direction
pad_direction = 0  # 0 for no movement, -1 for left, 1 for right

# Define pad attributes
pad_width = 60
pad_height = 10
pad_color = (255,255,255)  # Silver color

# Initialize pad position, velocity, and spacing
pad_x = (window_width - pad_width) // 2  # Start in the middle
pad_y = window_height - pad_height - 10  # Slightly above the bottom

# Define pad velocity
pad_velocity = 1 # Adjust as needed

# Define Pad Spacing:
spacing_from_border = 10

#-----------------------------------------------------------------Ball Initialization : -------------------------------------------------------------
# Define ball attributes
ball_radius = 4
ball_color = (255, 255, 255)  # White color

# Initialize ball position and velocity
ball_x = pad_x + pad_width // 2
ball_y = pad_y - ball_radius - 1 
ball_velocity_x = 0  # Initial horizontal velocity
ball_velocity_y = 0  # Initial vertical velocity

#---------------------------------------------------------------- Bricks Initialization: --------------------------------------------------------
# Define brick attributes
brick_width = 20
brick_height = 30

# Define brick layout dimensions
brick_rows = 4
brick_cols = window_width // brick_width

# Create a 2D list to represent the bricks
bricks = []

for row in range(brick_rows):
    brick_row = []
    for col in range(brick_cols):
        brick_x = col * brick_width
        brick_y = row * brick_height
        brick_row.append((brick_x, brick_y, brick_width, brick_height))
    bricks.append(brick_row)



# - -------------------------------------------------------   Score : ----------------------------------------------------------------------
# Load the highest score from a file
try:
    with open("highest_score.txt", "r") as file:
        highest_score = int(file.read())
except FileNotFoundError:
    highest_score = 0

# Font for displaying the score and highest score
font = pygame.font.Font(None, 36)

# ----------------------------------------------------------   Flags : -------------------------------------------------------------------------

# Initialize a flag to control ball movement
ball_moving = False

# Initialize a flag to handle game over state
game_over = False

# Initialize the score
score = 0

# --------------------------------------------------------  Music  : ---------------------------------------------------------------
# Load background music
bgm_sound = pygame.mixer.Sound("bgm_sound.mp3")

# Load game over music
game_over_sound = pygame.mixer.Sound("game_over.mp3")




def generate_brick_layout():
    bricks = []
    for row in range(brick_rows):
        brick_row = []
        for col in range(brick_cols):
            brick_x = col * brick_width
            brick_y = row * brick_height
            brick_color = (128, 128, 128)  # Gray color
            brick_row.append((brick_x, brick_y, brick_width, brick_height, brick_color))
        bricks.append(brick_row)
    return bricks
    
def reset_game():
    score = 0
    # Reset the pad and ball positions
    pad_x = (window_width - pad_width) // 2
    ball_x = pad_x + pad_width // 2
    ball_y = pad_y - ball_radius - 1
    # Regenerate the brick layout
    bricks = generate_brick_layout()


# ---------------------------------------------------------------- Main Game Loop : -----------------------------------------------------------------
# Main game loop
while True:
    bgm_sound.play(-1)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    
    
    # Handle user input to change the pad's movement direction
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        pad_direction = -1
    elif keys[pygame.K_RIGHT]:
        pad_direction = 1
    else:
        pad_direction = 0


    # Update the pad's position based on the direction and velocity
    pad_x += pad_direction * pad_velocity

    # Ensure the pad stays within the window boundaries
    pad_x = max(spacing_from_border, min(window_width - pad_width - spacing_from_border, pad_x))


    if game_over:
      if keys[pygame.K_SPACE]:
          # Reset the game when the space key is pressed
          game_over = False
          ball_x = pad_x + pad_width // 2
          ball_y = pad_y - ball_radius - 1
          ball_velocity_x = 0
          ball_velocity_y = 0
          ball_moving = False

    if not ball_moving:
        # Move the ball with the pad
        ball_x = pad_x + pad_width // 2
    
    if ball_moving:
        # -1 means play bgm in a loop  : 
        bgm_sound.play(-1)
        # Update the ball's position based on its velocity
        ball_x += ball_velocity_x
        ball_y += ball_velocity_y




        # Check for collisions with the window boundaries and reverse velocity if necessary
        if ball_x - ball_radius < 0 or ball_x + ball_radius > window_width:
            ball_velocity_x = -ball_velocity_x
        if ball_y - ball_radius < 0:
            ball_velocity_y = -ball_velocity_y



        # Check for collisions with the pad
        if (
            ball_y + ball_radius >= pad_y
            and ball_x + ball_radius >= pad_x
            and ball_x - ball_radius <= pad_x + pad_width
        ):
            ball_velocity_y = -ball_velocity_y


        # Check for game over condition (ball crosses the bottom)
        if ball_y + ball_radius > window_height:
            # Stop the background music
            bgm_sound.stop()
        
            # Play the game over sound
            game_over_sound.play()
            game_over = True
            ball_moving = False
                    
            # Update the highest score if needed
            if score > highest_score:
                highest_score = score
                with open("highest_score.txt", "w") as file:
                    file.write(str(highest_score))
        

        # Check for collisions with bricks
        for row in range(brick_rows):
            for col in range(brick_cols):
                brick_data = bricks[row][col]
                if brick_data:
                    brick_x, brick_y, brick_w, brick_h = brick_data
                    if (
                        ball_x + ball_radius >= brick_x
                        and ball_x - ball_radius <= brick_x + brick_w
                        and ball_y + ball_radius >= brick_y
                        and ball_y - ball_radius <= brick_y + brick_h
                    ):
                        # Reflect the ball's velocity
                        ball_velocity_y = -ball_velocity_y
                        # Remove the brick
                        bricks[row][col] = None
                        score += 1
                                             
                        # Remove the brick
                        bricks[row][col] = None
                        

    
    # Check for the Enter key press to initiate ball movement
    if not ball_moving and keys[pygame.K_RETURN]:
        
        # Randomly set the ball's initial direction
        
        # Set the Ball Velocity : 
        ball_velocity_x = random.choice([-0.3, 0.3])
        ball_velocity_y = -0.5
        ball_moving = True



    # Blit the background image onto the game window
    game_display.blit(background_image, (0, 0))


    # Draw the white pad: 
    pygame.draw.rect(game_display, pad_color, (pad_x, pad_y, pad_width, pad_height))


    # Draw the ball
    pygame.draw.circle(game_display, ball_color, (ball_x, ball_y), ball_radius)

    # Drawing the Bricks: 
    for row in range(brick_rows):
        for col in range(brick_cols):
            brick_data = bricks[row][col]
            if brick_data:
                brick_x, brick_y, brick_w, brick_h = brick_data
                pygame.draw.rect(game_display, (128, 128, 128), (brick_x, brick_y, brick_w, brick_h))
                pygame.draw.rect(game_display, (0, 0, 0), (brick_x, brick_y, brick_w, brick_h), 2)


    # Check and display the "Game Over" prompt
    if game_over:
        bgm_sound.stop()  # Stop the background music
        game_over_sound.play()  
        font = pygame.font.Font(None, 36)
        text = font.render("Game Over - Press Space to Restart", True, (255, 255, 255))
        text_rect = text.get_rect(center=(window_width // 2, window_height // 2))
        game_display.blit(text, text_rect)
        
        # Display the score and highest score
        score_text = font.render("Score: " + str(score), True, (255, 255, 255))
        highest_score_text = font.render("Highest Score: " + str(highest_score), True, (255, 255, 255))
        game_display.blit(score_text, (10, 10))
        game_display.blit(highest_score_text, (10, 50))
        reset_game()
        generate_brick_layout()



    # Update the display
    pygame.display.update()
       
            
```
